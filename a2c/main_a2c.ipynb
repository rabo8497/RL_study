a2c는 일반적으로 dqn보다 많은 에피소드를 이용하여 학습하지만, 학습 속도 면에서는 저 우월하다.

```python
import gym
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
import matplotlib.pyplot as plt
```

### Policy network와 Value network

```python
class PolicyNetwork(nn.Module):
    def __init__(self):
        super(PolicyNetwork, self).__init__()
        self.fc1 = nn.Linear(4, 128) 
        self.fc2 = nn.Linear(128, 2) 
        self.softmax = nn.Softmax(dim=1)     
    def forward(self, x):
        x = torch.relu(self.fc1(x))
        x = self.fc2(x)
        x = self.softmax(x)
        return x

class ValueNetwork(nn.Module):
    def __init__(self):
        super(ValueNetwork, self).__init__()
        self.fc1 = nn.Linear(4, 128) 
        self.fc2 = nn.Linear(128, 1) 
    def forward(self, x):
        x = torch.relu(self.fc1(x))
        x = self.fc2(x)
        return x
```

### Agent

```python
class Agent:
    def __init__(self, LR = 1e-3, GAMMA = 0.99) :
        self.device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
        self.LR = LR
        self.GAMMA = GAMMA
        self.device = device
        self.policy_net = PolicyNetwork().to(self.device)
        self.value_net = ValueNetwork().to(self.device)
        self.optimizer = optim.Adam(list(self.policy_net.parameters()) + list(self.value_net.parameters()), lr=LR)
        self.log_probs = []
        self.values = []
        self.rewards = []
    def select_action(self, state) :
        action_probabilities_tensor= self.policy_net(state) 
        action_probabilities_numpy= action_probabilities_tensor.detach().numpy()[0] 
        action= np.random.choice(2,p=action_probabilities_numpy) 
        return action
    def record(self, state, reward) :
        log_prob = torch.log(self.policy_net(state)[0][action])
        self.log_probs.append(log_prob)
        self.rewards.append(reward)
        value_state=self.value_net(state)
        self.values.append(value_state)
    def optimize(self) :
        discounted_rewards = []
        cumulative_reward = 0
        for reward in reversed(self.rewards):
            cumulative_reward = reward + self.GAMMA * cumulative_reward
            discounted_rewards.append(cumulative_reward)
        discounted_rewards.reverse()
        advantage = torch.tensor(discounted_rewards) - torch.cat(self.values)
        target_value = torch.tensor(discounted_rewards)
        value_loss = nn.MSELoss()(torch.cat(self.values), target_value.unsqueeze(1))
        policy_loss = []
        for log_prob, adv in zip(self.log_probs, advantage):
            policy_loss.append(-log_prob * adv)
        policy_loss = torch.cat(policy_loss).mean()
        self.optimizer.zero_grad()
        value_loss.backward(retain_graph=True)
        policy_loss.backward()
        self.optimizer.step()
    def initial(self) :
        self.log_probs = []
        self.values = []
        self.rewards = []
```

### 학습
def main으로 감쌌을 때 학습되지 않음 -> 원인 모름

```python
a2c_agent = Agent()
env = gym.make('CartPole-v1')
num_episodes = 2000
scores = []
for i in range(num_episodes):
    state = env.reset() 
    a2c_agent.initial()
    done = False 
    episode_reward = 0 
    # 최근 5개의 평균이 490점보다 크다면 종료
    # 이 방법보다는 학습 할때마다 10개의 샘플로 테스트하여 평균내어 비교하는 것이 적절하지만,
    # 시간이 오래 걸리기 때문에 아래의 방법으로 대체
    if sum(scores[-5:])/5 > 490 :
        save_path = "main_a2c_model.pt"
        print("달성, 저장 완료 (" + save_path + ")")
        torch.save(a2c_agent.policy_net.state_dict(), save_path)
        print("저장 완료 (" + save_path + ")")
        plt.plot(scores)
        plt.title('episode-score graph')
        plt.xlabel('episode')
        plt.ylabel('score')
        plt.savefig('main_a2c.png')
        plt.show()
        break
    while not done:
        state_tensor= torch.tensor(state).float().unsqueeze(0)  
        action = a2c_agent.select_action(state_tensor)
        next_state,reward,done,_= env.step(action) 
        episode_reward += reward
        next_state_tensor=torch.tensor(next_state).float().unsqueeze(0)
        a2c_agent.record(state_tensor, reward)
        if done:
            scores.append(episode_reward)
            a2c_agent.optimize()
            break
        else:
            state=next_state

    print(f"Episode {i}, Reward: {episode_reward}")
```

## 학습 결과 테스트

```python
import time
import torch
import torch.nn as nn
import torch.nn.functional as F
import gym

class PolicyNetwork(nn.Module):
    def __init__(self):
        super(PolicyNetwork, self).__init__()
        self.fc1 = nn.Linear(4, 128) 
        self.fc2 = nn.Linear(128, 2) 
        self.softmax = nn.Softmax(dim=1)     
    def forward(self, x):
        x = torch.relu(self.fc1(x))
        x = self.fc2(x)
        x = self.softmax(x)
        return x

def test_model(path = "main_a2c_model.pt") :
    env = gym.make('CartPole-v1', render_mode = 'human')
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    done = False
    score = 0
    save_path = "main_a2c_model.pt"
    model = PolicyNetwork().to(device)
    model.load_state_dict(torch.load(save_path))
    state = env.reset()
    while not done:
        env.render()
        state = torch.tensor(state, dtype=torch.float32, device=device).unsqueeze(0)
        time.sleep(0.02)
        action = model(state).max(1)[1]
        state, reward, done, _ = env.step(action.item())
        score += reward
    env.close()
    print('최종 점수: ' + str(score))
```

```python
test_model("main_a2c_model.pt")
```
