a2c는 일반적으로 dqn보다 많은 에피소드를 이용하여 학습하지만, 학습 속도 면에서는 저 우월하다.

```python
import gym
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
import matplotlib.pyplot as plt
```

### Policy network와 Value network

```python
class PolicyNetwork(nn.Module):
    def __init__(self):
        super(PolicyNetwork, self).__init__()
        self.fc1 = nn.Linear(4, 128) 
        self.fc2 = nn.Linear(128, 2) 
        self.softmax = nn.Softmax(dim=1)     
    def forward(self, x):
        x = torch.relu(self.fc1(x))
        x = self.fc2(x)
        x = self.softmax(x)
        return x

class ValueNetwork(nn.Module):
    def __init__(self):
        super(ValueNetwork, self).__init__()
        self.fc1 = nn.Linear(4, 128) 
        self.fc2 = nn.Linear(128, 1) 
    def forward(self, x):
        x = torch.relu(self.fc1(x))
        x = self.fc2(x)
        return x
```

### Agent

```python
learning_rate = 0.001
gamma = 0.99 
num_episodes = 2000

env = gym.make('CartPole-v1')

policy_net = PolicyNetwork()
value_net = ValueNetwork()
optimizer = optim.Adam(list(policy_net.parameters()) + list(value_net.parameters()), lr=learning_rate)

scores = []
for i in range(num_episodes):
    state = env.reset() 
    done = False 
    episode_reward = 0 
    log_probs = []
    values = []
    rewards = []
    # 최근 10개의 평균이 450점보다 크다면 종료
    if sum(scores[-10:])/10 > 450 :
        save_path = "main_a2c_model.pt"
        print("달성, 저장 완료 (" + save_path + ")")
        torch.save(policy_net.state_dict(), save_path)
        print("저장 완료 (" + save_path + ")")
        plt.plot(scores)
        plt.title('episode-score graph')
        plt.xlabel('episode')
        plt.ylabel('score')
        plt.savefig('main_a2c.png')
        plt.show()
        break
    while not done:
        state_tensor= torch.tensor(state).float().unsqueeze(0) 
        action_probabilities_tensor= policy_net(state_tensor) 
        action_probabilities_numpy= action_probabilities_tensor.detach().numpy()[0] 
        action= np.random.choice(env.action_space.n,p=action_probabilities_numpy) 
        next_state,reward,done,_= env.step(action) 
        log_prob = torch.log(action_probabilities_tensor[0][action])
        episode_reward += reward
        rewards.append(reward)
        log_probs.append(log_prob)
        next_state_tensor=torch.tensor(next_state).float().unsqueeze(0)
        value_state=value_net(state_tensor)
        values.append(value_state)
        
        if done:
            scores.append(episode_reward)
            discounted_rewards = []
            cumulative_reward = 0
            for reward in reversed(rewards):
                cumulative_reward = reward + gamma * cumulative_reward
                discounted_rewards.append(cumulative_reward)
            discounted_rewards.reverse()
            advantage = torch.tensor(discounted_rewards) - torch.cat(values)
            
            target_value = torch.tensor(discounted_rewards)
            value_loss = nn.MSELoss()(torch.cat(values), target_value.unsqueeze(1))
            
            policy_loss = []
            for log_prob, adv in zip(log_probs, advantage):
                policy_loss.append(-log_prob * adv)
                
            policy_loss = torch.cat(policy_loss).mean()

            optimizer.zero_grad()
            value_loss.backward(retain_graph=True)
            policy_loss.backward()
            optimizer.step()
            
            break
            
        else:
            state=next_state
            
    print(f"Episode {i}, Reward: {episode_reward}")
```

## 학습 결과 테스트

```python
import time
import torch
import torch.nn as nn
import torch.nn.functional as F
import gym

class PolicyNetwork(nn.Module):
    def __init__(self):
        super(PolicyNetwork, self).__init__()
        self.fc1 = nn.Linear(4, 128) 
        self.fc2 = nn.Linear(128, 2) 
        self.softmax = nn.Softmax(dim=1)     
    def forward(self, x):
        x = torch.relu(self.fc1(x))
        x = self.fc2(x)
        x = self.softmax(x)
        return x

def test_model(path = "main_a2c_model.pt") :
    env = gym.make('CartPole-v1', render_mode = 'human')
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    done = False
    score = 0
    save_path = "main_a2c_model.pt"
    model = PolicyNetwork().to(device)
    model.load_state_dict(torch.load(save_path))
    state = env.reset()
    while not done:
        env.render()
        state = torch.tensor(state, dtype=torch.float32, device=device).unsqueeze(0)
        time.sleep(0.02)
        action = model(state).max(1)[1]
        state, reward, done, _ = env.step(action.item())
        score += reward
    env.close()
    print('최종 점수: ' + str(score))
```

```python
test_model("main_a2c_model.pt")
```
